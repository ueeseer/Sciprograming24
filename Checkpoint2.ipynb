```python
import cmath

def calculate_discriminant(a, b, c):
    return b**2 - 4*a*c

def main():
    # Read in the coefficients
    a = float(input("Enter coefficient a: "))
    b = float(input("Enter coefficient b: "))
    c = float(input("Enter coefficient c: "))

    # Calculate the discriminant
    d = calculate_discriminant(a, b, c)

    # Calculate the roots based on the discriminant
    if a == 0:
        if b != 0:
            root = -c / b
            print(f"We got Linear equation root: {root}")
        else:
            print(f"Invalid equation. Coefficients a and b cannot both be zero.")
    else:
        if d > 0:
            root1 = (-b + cmath.sqrt(d)) / (2*a)
            root2 = (-b - cmath.sqrt(d)) / (2*a)
            print(f"Two real roots: {root1.real} and {root2.real}")
        elif d == 0:
            root = -b / (2*a)
            print(f"Single real root: {root}")
        else:
            root1 = (-b + cmath.sqrt(d)) / (2*a)
            root2 = (-b - cmath.sqrt(d)) / (2*a)
            print(f"Two complex roots: {root1} and {root2}")

# Run the main function
main()
```
